```python
palabra_secreta = 'cueva'
palabra_ingresada = ''
lista_emoji = []
i=0
longitud=0 
indice = 0 

print('*************************************')
print('                WORDLE               ')
print('*************************************')
      
print('')
print('ESPECIFICACIONES')
print('')
print('🟢 Letra en la ubicación correcta')
print('🟡 Letra existente pero no en su ubicacacion correcta')
print('🔘 Letra existenteno exitente en la palabra')
print('')

#Función para verificar los aciertos
def verifica(palabra1, palabra2):
    for i in range(5):
        if palabra1[i] == palabra2[i]:

            lista_resultado.append('['+ palabra2[i]+']')
            lista_emoji.append('🟢')

           
        elif palabra2[i] in  palabra1:

            lista_resultado.append('('+ palabra2[i]+')')
            lista_emoji.append('🟡')
            
        else:

            lista_resultado.append(palabra2[i])
            lista_emoji.append('🔘')
        
# for para 6 intentos
for indice in range(6):
    
    #Validación -- Que se ingrese una palabra de 5 letras
    while True:
        palabra_ingresada=input('Ingrese una palabras con 5 letras: ')
        longitud=len(palabra_ingresada)
        if longitud == 5:
            break
        else:
            print('Error! La palabra debe tener 5 letras')

    #Intentos conteo
    print(f'Intento N°{indice+1}')
    lista_resultado = []
    lista_emoji = []
    #Llamamos verificación
    verifica(palabra_secreta, palabra_ingresada)

    print(lista_resultado)
    print(lista_emoji)
    print('')

    #si pierde
    if palabra_secreta == palabra_ingresada:
        print('Felicidaes. Has ganado!')
        break
    

#si pierde
if palabra_secreta != palabra_ingresada:
        print('Lo lamento. Has perdido!')
```
