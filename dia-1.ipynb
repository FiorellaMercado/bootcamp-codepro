```python
#solicitar datos
total_caramelos = int(input('Ingrese cantidad de caramelos'))
cantidad_ninos= int(input('Ingrese cantidad de niños')) 

#calculos
caramelos_por_nino= total_caramelos//cantidad_ninos
caramelos_restantes = total_caramelos % cantidad_ninos

#resultados
print(f'La cantidad de caramelos por niño es: {caramelos_por_nino}')
print(f'La cantidad de caramelos sobrantes es: {caramelos_restantes}')
```
