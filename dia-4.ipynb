```python
tarjeta_personal={}

def crear():
    clave=input("Ingrese el nombre de la clave")
    valor=input("Ingrese un valor para la clave")
    tarjeta_personal[clave]=valor


def modificar():
    clave=input("Ingrese el nombre de la clave a modificar")
    valor=input("Ingrese un nuevo valor")
    tarjeta_personal[clave]=valor

def eliminar():
    clave= input("Ingrese el nombre de la clave que desea eliminar")
    del tarjeta_personal[clave]


while True:
    opcion=int(input("Ingrese una opcion: 1. Crear, 2. Modificar, 3. Eliminar, 4. Salir"))
    if opcion == 1:
        crear()
        print(tarjeta_personal)
    elif opcion ==2:
        modificar()
        print(tarjeta_personal)
    elif opcion==3:
        eliminar()
        print(tarjeta_personal)
    elif opcion == 4:
        break 
    else :
        print("Opción no válida")
       
        
        

```

```python
agenda={}

def crear():
    nombre=input("Ingrese el nombre del contacto")
    telefono=input("Ingrese el telefono")
    correo=input("Ingrese el correo")

    agenda[nombre] = {"telefono":telefono, "correo":correo}
    
def modificar():
    contacto_modificar=input("Ingrese el nombre del contacto a modificar")
    telefono=input("Ingrese el telefono")
    correo=input("Ingrese el correo")

    agenda[contacto_modificar] = {"telefono":telefono, "correo":correo}



while True:
    
    opcion=int(input("Ingrese una opcion: 1. Mostrar, 2. Crear, 3. Modificar, 4. Salir"))
    if opcion == 1:
        print(agenda)
    elif opcion ==2:
        crear()       
    elif opcion==3:
        print("Cuál desea modificar:")
        for clave in agenda:
            print("- ",clave)
        modificar()

    elif opcion == 4:
        break 
    else :
        print("Opción no válida")




```

```python
class Animal:
    especie = "ave"
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

animal1 = Animal("águila","blanca")
animal2 = Animal("pájaro carpintero","gris")

print(f"Me llamo: {animal1.nombre} y soy un ave de color: {animal1.color}")
print(f"Me llamo: {animal2.nombre} y soy un ave de color: {animal2.color}")
```

```python
class Animal:
    especie = "mamíferos"
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    
    def onomatopeya(self, sonido):
        self.sonido=sonido
    def mueve(self,movimiento):
        self.movimiento=movimiento

animal1 = Animal("vaca","blanca")  
animal1.onomatopeya("muuu")
animal1.mueve("pasos")

print(f"Cuando el/la {animal1.nombre} suena hace {animal1.sonido} y cuando me muevo hago {animal1.movimiento}") 

animal2 = Animal("perro","negro")
animal2.onomatopeya("guau")
animal2.mueve("pasos")

print(f"Cuando el/la {animal2.nombre} suena hace {animal2.sonido} y cuando me muevo hago {animal2.movimiento}") 
```

```python
class Utiles:
    def __init__(self, nombre, precio):
        self.nombre=nombre
        self.precio=precio
    

class Lapices(Utiles):
    def detalle(self):
        print(f"{self.nombre} a {self.precio}")

class Borradores(Utiles):
    def detalle(self):
        print(f"{self.nombre} a {self.precio}")


lapiz_de_papel=Lapices("Lapiz de papel HB 2", 5000)
lapiz_de_papel.detalle()

goma_de_borra=Borradores("Goma de borrar", 3000)
goma_de_borra.detalle()
    
```

```python
class Personajes:
    def __init__(self, nombre):
        self.nombre=nombre
    

class Heroe(Personajes):
    def correr(self):
        print(f"Soy {self.nombre} y corro super rápido")

class Villano(Personajes):
    def llamas(self):
        print(f"Soy {self.nombre} y lanzo llamas")


hereo1=Heroe("Flash")
hereo1.correr()

villano2=Villano("Dragon")
villano2.llamas()
```
