```python
temperatura_ingresada = int(input('Ingrese la temperatura en grados Celsius'))

print(f'La teperatura es {temperatura_ingresada} ')

def clasifica_temepratura(temperatura):

    if temperatura < -10 or temperatura > 45:
        print("Ñamano mba'e")
    elif temperatura <= 13:
        print("Ro'y koko")
    elif temperatura <= 23:
        print("Refresca'i")
    elif temperatura <= 28:
        print("Oikoite")
    elif temperatura <= 45:
        print("Hakurooooo")

clasifica_temepratura(temperatura_ingresada)
```

```python
cont=0
frase = input('Ingrese una frase')
for caracter in frase:
    if caracter == 'o' or caracter == 'O':
        cont=cont+1
        print("Letra 'o' detectada")

print(f"Hay {cont} letras 'o' en la frase")
```
